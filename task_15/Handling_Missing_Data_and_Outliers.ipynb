```python
import numpy as np
import pandas as pd
```

# Handling Missing Data

```python
data = {'Name': ['Usama','Zaryab','Abtuha'],
     'Age':[21,np.nan,20],
     'Address': ['Rawalpindi',np.nan,'Islamabad'] }

df = pd.DataFrame(data)

df
```

```python
df.isnull()
```

```python
df.notnull()
```

# Filtering and Replacing Missing Data

```python
'''Filtering / Dropping Missing Data'''
```

```python
# Dropping Missing Values

df.dropna()
```

```python
data = pd.DataFrame([[1., 6.5, 3.], [1., np.nan, np.nan],
                      [np.nan, np.nan, np.nan], [np.nan, 6.5, 3.]])

print('Data before Cleaning....')
print(data.head())

print('Data after Cleaning....')
data.dropna(axis=0)
```

```python
data.dropna(axis=0 , how='all')
```

```python
'''Filling / Replacing Missing Data'''
```

```python
df = pd.DataFrame(np.random.randn(7, 3))

df.iloc[:4, 1] = np.nan
df.iloc[:2, 2] = np.nan

df
```

```python
df.fillna(0) #Filling out missing Data with 0
```

```python
# Filling different columns with different values
df.fillna({1:0.7 , 2: 2.3}) 
```

```python
df = pd.DataFrame(np.random.randn(6, 3))

df.iloc[2:, 1] = np.nan
df.iloc[4:, 2] = np.nan

df
```

```python
df.fillna(method='ffill')
```

```python
df.fillna(df.mean())
```

# Detecting and Removing Outliers

```python
df = pd.DataFrame(np.random.randn(1000, 4))
df.head()
```

```python
df.describe()
```

```python
# Checking for values exceeding 3 or -3
df[(np.abs(df)>3).any(1)] 
```

```python
# Dealing with outliers exceeding 3 or -3

df[np.abs(df)>3] = np.sign(df)*3

df[(np.abs(df)>3).any(1)] 
```

```python
df.describe()
```

```python
# Scaling data between 1 nad -1

df = np.sign(df)

df
```
