# Concatenation DataFrames

# Merging DataFrames

```python
registrations = pd.DataFrame({'reg_id':[1,2,3,4],'name':['Andrew','Bobo','Claire','David']})
logins = pd.DataFrame({'log_id':[1,2,3,4],'name':['Xavier','Andrew','Yolanda','Bobo']})
```

```python
registrations
```

```python
logins
```

## Inner Join

**Match up where the key is present in BOTH tables**

```python
pd.merge(registrations,logins, how='inner')
```

```python
'''In Inner Join order provided does not matter'''
```

```python
pd.merge(logins, registrations, how='inner',on='name')
```

## Left Merge

**Match up AND include all rows from Left Table.**

```python
pd.merge(left=registrations, right=logins, how='left', on='name')
```

## Right Merge

**Match up AND include all rows from Right Table**

```python
pd.merge(left=registrations, right=logins, how='right', on='name')
```

## Outer Merge

**Match up on all info found in either Left or Right Table**

```python
pd.merge(logins, registrations, how='outer', on='name')
```

**Merging on Index**

```python
registrations =registrations.set_index('name')

registrations
```

```python
pd.merge(registrations, logins, left_index=True, right_on='name',how='inner')
```

```python
registrations =registrations.reset_index()
```

```python
registrations
```

```python
registrations.columns = ['reg_name','reg_id']
```

```python
registrations
```

```python
pd.merge(registrations,logins,how='inner',left_on='reg_name',right_on='name')
```

```python
pd.merge(registrations,logins,left_on='reg_name',right_on='name').drop('reg_name',axis=1)
```

```python
registrations.columns = ['name','id']
```

```python
logins.columns = ['id','name']
```

```python
registrations
```

```python
logins
```

```python
pd.merge(registrations,logins,on='name')
```

```python
pd.merge(registrations,logins,on='name',suffixes=('_reg','_log'))
```
